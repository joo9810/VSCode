선 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-9, 10)
y = x ** 2

plt.plot(x, y,
         linestyle = ":",
         marker = "*",
         markersize = 10,
         markerfacecolor = 'red',
         markeredgecolor = 'green')
plt.show()
```

막대 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-9, 10)
y = x ** 2

plt.bar(x, y)
plt.show()
```

누적 막대 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)

a = np.random.rand(10) # 0 ~ 1 사이의 랜덤 실수 10개
b = np.random.rand(10)
c = np.random.rand(10)

summary = [a, b, c]

for i in range(0, 3):
    plt.bar(x, summary[i], bottom = np.sum(summary[:i], axis = 0)) # 0 ~ i-1 인덱스까지 열을 기준으로 더하기
plt.show()

# a, b, c는 각각 1 x 10 데이터이므로 summary는 3 x 10 데이터
# 따라서 열을 기준으로 더하면 각 3개의 값을 합친 1 x 10 데이터가 나온다
```

산점도 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(10)
y = np.random.rand(10)
color = np.random.randint(0, 100, 10)
size = np.pi * 1000 * np.random.rand(10)

plt.scatter(x, y, c = color, s = size, alpha = 0.5) # alpha는 투명도인듯
plt.show()
```
