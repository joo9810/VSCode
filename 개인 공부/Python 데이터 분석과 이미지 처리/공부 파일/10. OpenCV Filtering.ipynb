직접 커널을 생성하여 필터 적용하기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("gray_image.jpg")

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

size = 4
kernal = np.ones((size, size), np.float32) / (size ** 2) # 1/16으로 채워진 4행 4열짜리 행렬
print(kernal)

dst = cv2.filter2D(image, -1, kernal) # -1: 출력 이미지와 입력 이미지의 데이터 타입을 동일하게 유지
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

Basic Blurring

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('gray_image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

dst = cv2.blur(image, (4, 4)) # 위에서와 같이 size를 4로 잡는다는 의미
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('gray_image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

# kernal size = 홀수
dst = cv2.GaussianBlur(image, (7, 7), 0) # 중심 필터가 가운데에 와야 하기 때문에 홀수로 지정해야 함, 0: 표준편차
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
