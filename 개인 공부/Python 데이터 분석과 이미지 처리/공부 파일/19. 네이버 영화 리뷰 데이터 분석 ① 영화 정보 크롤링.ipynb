```python
import requests
from bs4 import BeautifulSoup
import textwrap
import numpy as np
import pandas as pd
```

```python
url = "https://www.netflix.com/kr/title/70024111"

def append_text(data):
    list_text = []
    for i in data:
        list_text.append(i.text)
    return list_text

def movie_info(url):
    request = requests.get(url)

    html = request.text.strip()

    soup = BeautifulSoup(html, 'html.parser')
    title = soup.find('div', class_='title-info').find('h1').text
    synopsis = soup.find('div', class_='title-info-synopsis').text
    talent = soup.find('span', class_='title-data-info-item-list').text
    genre = soup.select('span > a') # 장르가 하나가 아니라서 <span>안에 <a>가 2개 있어서 select로 가져와야 함
    mood = soup.select('span.more-details-item.item-mood-tag')
    cast = soup.select('span.more-details-item.item-cast')
    
    hook = soup.find('div', class_='hook-text')
    if hook: # hook의 값이 존재하면 True
        hook = hook.text
    else: # hook의 값이 없으면 None이므로 False
        hook = None

    print("제목:", title)
    print("\n시놉시스:", textwrap.fill(synopsis))
    print("\n홍보 문구:", hook)
    print("\n주연:", talent)
    print("\n장르:", append_text(genre))
    print("\n영화 특징:", append_text(mood))
    print("\n출연:", append_text(cast))
```

```python
# 올드보이
movie_info('https://www.netflix.com/kr/title/70024111')
```

```python
# 럭키
movie_info('https://www.netflix.com/kr/title/80149102')
```
