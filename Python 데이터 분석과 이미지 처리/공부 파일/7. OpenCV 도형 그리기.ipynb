직선 그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((1024, 1024, 3), 255, np.uint8)
image = cv2.line(image, (0, 0), (512, 512), (255, 0, 0), 3) # (0, 0)에서 (512, 512)까지 Red = 255인 선을 굵기 3으로 그리기

plt.imshow(image)
plt.show()
```

사각형 그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((1024, 1024, 3), 255, np.uint8)
image = cv2.rectangle(image, (200, 200), (800, 800), (255, 0, 0), 5)

plt.imshow(image)
plt.show()
```

원 그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((1024, 1024, 3), 255, np.uint8)
image = cv2.circle(image, (500, 500), 200, (255, 0, 0), 10)

plt.imshow(image)
plt.show()
```

다각형 그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((1024, 1024, 3), 255, np.uint8)
point = np.array([[100, 100], [800, 400], [900, 700], [200, 800]])
image = cv2.polylines(image, [point], True, (0, 0, 255), 5)

plt.imshow(image)
plt.show()
```

텍스트 그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((1024, 1024, 3), 255, np.uint8)
image = cv2.putText(image, "HelloWorld!", (100, 500), cv2.FONT_ITALIC, 5, (255, 0, 0), 5)

plt.imshow(image)
plt.show()
```
