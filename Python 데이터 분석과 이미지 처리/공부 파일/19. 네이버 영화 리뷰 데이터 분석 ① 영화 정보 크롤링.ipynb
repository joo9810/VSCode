```python
import requests
from bs4 import BeautifulSoup
import textwrap
import numpy as np
import pandas as pd
```

```python
url = "https://www.netflix.com/kr/title/70024111"

def append_text(data):
    list_text = []
    for i in data:
        list_text.append(i.text)
    return list_text

def movie_info(url):
    request = requests.get(url)

    html = request.text.strip()

    soup = BeautifulSoup(html, 'html.parser')
    title = soup.find('div', class_='title-info').find('h1').text
    synopsis = soup.find('div', class_='title-info-synopsis').text
    talent = soup.find('span', class_='title-data-info-item-list').text
    genre = soup.select('span > a') #    <span> <a> 2  select  
    mood = soup.select('span.more-details-item.item-mood-tag')
    cast = soup.select('span.more-details-item.item-cast')
    
    hook = soup.find('div', class_='hook-text')
    if hook: # hook   True
        hook = hook.text
    else: # hook   None False
        hook = None

    print(":", title)
    print("\n:", textwrap.fill(synopsis))
    print("\n :", hook)
    print("\n:", talent)
    print("\n:", append_text(genre))
    print("\n :", append_text(mood))
    print("\n:", append_text(cast))
```

```python
# 
movie_info('https://www.netflix.com/kr/title/70024111')
```
