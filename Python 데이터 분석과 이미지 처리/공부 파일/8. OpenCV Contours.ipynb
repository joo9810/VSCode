Contours

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('gray_image.jpg')

plt.imshow(image)
plt.show()

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB))
plt.show()

contours = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[0] # 컨투어는 하얀색의 경계를 찾는다고 생각하면 될듯?
# RETR_TREE: 윤곽선을 찾는 방법, CHAIN_APPROX_SIMPLE: 윤곽선을 근사화하는 방법, [0]: 윤곽선 정보 (OpenCV 버전에 따라 0으로 할지 1로 할지 다름)
image = cv2.drawContours(image, contours, -1, (255, 0, 0), 3) # -1: 모든 컨투어를 가져온다

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
