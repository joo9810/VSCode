간단한 직선 그래프 그리기

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [1, 2, 3]

plt.plot(x, y)
plt.title('Y = X Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

그래프 저장하기 ②

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.pi * 10, 500) # 0 ~ 31.4 사이에 500개의 점을 균일하게 찍기

fig, axes = plt.subplots(2, 1)
axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))
plt.show()

fig.savefig('sin&cos.png')
```

선 그래프 그리기 ①

```python
x = np.arange(-9, 10)
y = x ** 2

plt.plot(x, y, linestyle = ':', marker = '*')
plt.show()
```

선 그래프 그리기 ②

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-9, 10)
y1 = x ** 2
y2 = -x

plt.plot(x, y1, linestyle = ':', marker = '*', color = 'red', label = 'y = x * x')
plt.plot(x, y2, label = 'y = -x')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(shadow = True, borderpad = 1) # borderpad: 범례와 내용 사이의 간격
plt.show()
```
