특정 웹 사이트 HTML 코드 추출 ②

```python
import requests
from bs4 import BeautifulSoup

request = requests.get("http://www.dowellcomputer.com/main.jsp")
html = request.text.strip()

# html 소스코드를 BeautifulSoup 객체로 변환
soup = BeautifulSoup(html, 'html.parser') # 추가적인 설치 없이 사용할 수 있는 기본 파서 ('lxml', 'html5lib'도 있다고 함)

print(soup) # html 소스코드에서 왼쪽 공백을 모두 없애서 나타낸듯
```

```python
import numpy as np

links = soup.select('td > a')

print(np.shape(links)) # links는 5개의 요소를 가진 3차원 배열

print(links) # <td> 태그의 바로 아래에 있는 <a> 태그들을 선택하여 이들을 리스트로 반환
# <a href="./study/study.jsp?studyID=85"><b>C언어 기초 프로그래밍 강좌 20강 - 동적 메모리의 활용 (C Programming Tutorial For Beginners 2017 #20) </b></a>
# <a href="./study/study.jsp?studyID=84"><b>C언어 기초 프로그래밍 강좌 19강 - 동적 메모리 (C Programming Tutorial For Beginners 2017 #19) </b></a>
# <a href="./study/study.jsp?studyID=83"><b>C언어 기초 프로그래밍 강좌 18강 - 파일 입출력 (C Programming Tutorial For Beginners 2017 #18) </b></a>
# <a href="./study/study.jsp?studyID=82"><b>C언어 기초 프로그래밍 강좌 17강 - 구조체의 활용 ② (C Programming Tutorial For Beginners 2017 #17) </b></a>
# <a href="./study/study.jsp?studyID=81"><b>C언어 기초 프로그래밍 강좌 16강 - 구조체의 활용 ① (C Programming Tutorial For Beginners 2017 #16) </b></a>
```

```python
for i in np.arange(5):
    print(links[i].has_attr('href'))
# 5개의 <a> 태그 모두 href 속성을 가지고 있다.
```

```python
for i in np.arange(5):
    print(links[i].get('href').find('study')) # study값이 존재하면 그 개수만큼, 없으면 -1을 출력
# 5개의 <a> 태그 모두 study라는 단어를 2개씩 가지고 있다.
```

```python
for i in links:
    if i.has_attr('href'): # 소스코드가 href 속성을 가지고 있다면
        if i.get('href').find('study') != -1: # links의 각 <a> 태그들이 study를 가지고 있는지 확인
         print(i.text) #  태그 안에 있는 모든 텍스트를 추출
```
