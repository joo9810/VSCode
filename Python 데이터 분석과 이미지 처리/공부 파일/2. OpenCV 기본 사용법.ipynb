이미지 출력

```python
import cv2
import matplotlib.pyplot as plt

img_basic1 = cv2.imread('cat.jpg', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic1, cv2.COLOR_BGR2RGB))
plt.show() # BGR을 RGB로

img_basic2 = cv2.cvtColor(img_basic1, cv2.COLOR_BGR2GRAY)
plt.imshow(img_basic2)
plt.show() # BGR을 GRAY로

img_basic3 = cv2.cvtColor(img_basic2, cv2.COLOR_GRAY2RGB)
plt.imshow(img_basic3)
plt.show() # GRAY를 RGB로

```
