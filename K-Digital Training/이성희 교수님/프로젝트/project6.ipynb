## 등분산 검정

>귀무가설 : 두 집단의 분산이 같다  
>대립가설 : 두 집단의 분산이 다르다

```python
from scipy.stats import bartlett
import numpy as np

bartlett(red_density, white_density)
```

#### Welch's T-test (등분산 가정을 만족하지 않더라도 사용 가능)

>귀무가설 : 레드와인과 화이트와인의 무게감 차이가 없다  
>대립가설 : 레드와인과 화이트와인의 무게감 차이가 있다

```python
import scipy.stats as stats
```

```python
t_stat, p_value = stats.ttest_ind(red_density, white_density, equal_var = False)
t_stat, p_value
```

상관계수가 가장 높은 변수가 (무게감, 알콜)  
두번째로 높은 변수가 (무게감, 잔여 설탕)  
따라서 다중공선성 문제를 제거하기 위해 무게감 컬럼을 제거하기로 결정
