```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def scraping_use_find(html):
    soup = BeautifulSoup(html.read(), 'html.parser')

    period_name_list = []
    for i in soup.find_all('p', {'class': 'period-name'}):
        period_name_list.append(i.string)

    short_desc_list = []
    for i in soup.find_all('p', {'class': 'short-desc'}):
        short_desc_list.append(i.string)

    temp_list = []
    for i in soup.find_all('p', {'class': 'temp'}):
        temp_list.append(i.string)

    image_list = []
    soup2 = soup.find('ul', {'id': 'seven-day-forecast-list'})

    for i in soup2.find_all('img'):
        image_list.append(i['alt'])

    print('[find 함수 사용]')
    print(f'총 tomestone-container 검색 개수 : {len(period_name_list)}')
    for i in range(len(period_name_list)):
        print(f'[Period]: {period_name_list[i]}\n'
            f'[Short desc]: {short_desc_list[i]}\n'
            f'[Temperature]: {temp_list[i]}\n'
            f'[Image desc]: {image_list[i]}')
        print('-' * 100)
```

```python
url = 'https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168'
html = urlopen(url)

scraping_use_find(html)
```

```python
def scraping_use_select(html):
    soup = BeautifulSoup(html.read(), 'html.parser')

    period_name_list2 = []
    for i in soup.select('p.period-name'):
        period_name_list2.append(i.string)

    short_desc_list2 = []
    for i in soup.select('p.short-desc'):
        short_desc_list2.append(i.string)

    temp_list2 = []
    for i in soup.select('p.temp'):
        temp_list2.append(i.string)

    image_list2 = []

    for i in soup.select('#seven-day-forecast-list img'):
        image_list2.append(i['alt'])

    print('[select 함수 사용]')
    print(f'총 tomestone-container 검색 개수 : {len(period_name_list2)}')
    for i in range(len(period_name_list2)):
        print(f'[Period]: {period_name_list2[i]}\n'
            f'[Short desc]: {short_desc_list2[i]}\n'
            f'[Temperature]: {temp_list2[i]}\n'
            f'[Image desc]: {image_list2[i]}')
        print('-' * 100)
```

```python
url = 'https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168'
html = urlopen(url)

scraping_use_select(html)
```
