[1] for문 하나만 쓰고 구구단 전부 출력

```python
for i in range(20, 100):
    if i % 10 == 0:
        print(f'{"-" * 5} {int(i / 10)}단 {"-" * 5}')
    else:
        print(f'{i // 10} * {i % 10} = {(i // 10) * (i % 10)}')
```

[2] 구구단 2줄로 출력

```python
for i in range(1, 19):
    if i == 1:
        for l in range(2, 6):
            print(f'----{l}단----', end = '  ')
    elif i == 10:
        print('\n')
        for l in range(2, 6):
            print(f'----{l+4}단----', end = '  ')        
    print('')
    for j in range(2, 6):
        if i < 10: # i가 1 ~ 9까지
            if i % 9 == 0: # i가 9의 배수일때
                if i == 9: # i가 9일때
                    print(f'{j} * {i} = {j * i:2}', end='   ')
                else: # i가 18일때
                    print(f'{j} * {int(i/2)} = {j * int(i/2):2}', end='   ')
            else:
                print(f'{j} * {i%9} = {j * i:2}', end='   ')
        else: # i가 10 ~ 18까지
            if i % 9 == 0: # i를 9로 나눈 나머지가 0일때
                if i == 9: # i가 9일때
                    print(f'{j+4} * {i} = {j+4 * i:2}', end='   ')
                else: # i가 18일때
                    print(f'{j+4} * {int(i/2)} = {(j+4) * int(i/2):2}', end='   ')
            else: # i가 9의 배수가 아닐때
                print(f'{j+4} * {i%9} = {(j+4) * (i%9):2}', end='   ')
```

```python
def print_gugudan():
    for i in range(2, 10, 4):
        for j in range(10):
            if j == 0:
                for k in range(i, i + 4):
                    if k < 10:
                        print(f"----{k}단----", end="  ")
                print()
            else:
                for k in range(i, i + 4):
                    if k < 10:
                        print(f"{k} * {j} = {k * j:2}", end="   ")
                print()
        print()

print_gugudan()
```
