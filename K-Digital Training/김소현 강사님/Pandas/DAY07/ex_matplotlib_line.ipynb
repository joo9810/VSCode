### [ Matplotlib 활용한 시각화 ]
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# [1] 모듈 로딩
# - 패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plot(y축 데이터)
plt.plot([1, 2, 3, 4])
plt.xlabel('data')
plt.ylabel('result')
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(X축 데이터, Y축 데이터, fmf 포맷 스트링)
# fmf 포맷 스트링 : 그래프 색상, 두께, 스타일 등 지정
plt.plot([10, 20, 30], [5, 10, 15], 'ro--')
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(X축 데이터, Y축 데이터, line2D 속성으로 설정)
# fmf 포맷 스트링 : 그래프 색상, 두께, 스타일 등 지정
# fmf 포맷스트링 ==> Line2D의 속성으로 설정
# - color = 'red', marker = 'o', linestyle ] 'dashed'
# - markersize = 10, linewidth = 4
# - markeredgecolor or mec, markeredgewidth or mew
plt.plot([10, 20, 30], [5, 10, 15], color = 'red', marker = 'o', linestyle = 'dashed',
         markersize = 10, linewidth = 4, mec = 'blue', mew = 3)
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plt.plot([10, 20, 30], [15, 25, 35], 'bo-')
# plt.plot([10, 20, 30], [18, 29, 25], 'ro--')
plt.plot([10, 20, 30], [15, 25, 35], 'bo-', [10, 20, 30], [18, 29, 25], 'ro--')
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - x축, y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성값을 범례에서 사용함
plt.plot([10, 20, 30], [15, 25, 20], 'bo--', label = 'V1')
plt.plot([10, 20, 30], [5, 30, 28], 'ro--', label = 'V2')
plt.title('[MPLIB LINE2D]')
plt.xlabel('Data')
plt.ylabel('Result')
plt.legend()
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - x축, y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성값을 범례에서 사용함
#                              loc 속성으로 위치 지정 가능 [기본값: best]
plt.plot([10, 20, 30], [15, 25, 20], 'bo--', label = 'V1')
plt.plot([10, 20, 30], [5, 30, 28], 'ro--', label = 'V2')
plt.title('[MPLIB LINE2D]')
plt.xlabel('Data')
plt.ylabel('Result')
plt.legend(loc = 'upper center')
plt.show()
```
