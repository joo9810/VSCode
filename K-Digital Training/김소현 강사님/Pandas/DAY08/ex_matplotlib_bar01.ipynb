```python
# [1-2] 데이터 준비
F_DATA = '/Users/anhyojun/VSCode/K-Digital Training/Pandas/숙제/시도별 전출입 인구수.xlsx'
```

```python
# [1-3] DATA ==> DataFrame 형태 저장
dataDF = pd.read_excel(F_DATA)
```

- [3] 데이터 전처리
    * 기본확인 파악된 내용들 처리
    * 불필요 데이터 삭제 : 0번행 삭제
    * 결측치 데이터 처리 : '전출지별' 컬럼 처리
    * 이상치 데이터 처리 : '-'에 대한 처리
    * 데이터 자료형 일치 : 형 변환

```python
# [3-1] 불필요 데이터 처리
copyDF = dataDF.drop(index = 0)
copyDF.head()
```

```python
# [3-2] 결측치 데이터 처리 => isna()/isnull()
copyDF.isna().sum()
```

```python
a = copyDF.loc[:, ['1984', '1985', '1986']]
a
```

```python
dataDF.columns
```

```python
a[a['1984'].isna() | a['1985'].isna() | a['1986'].isna()]
```

```python
a[a.isna().T.sum() > 0]
```

```python
# - '전출지별'
copyDF['전출지별'] = copyDF['전출지별'].ffill()
copyDF
```

```python
copyDF[['1984', '1985', '1986']] = copyDF[['1984', '1985', '1986']].fillna(0)
copyDF.isna().sum()
```

```python
# [3-3] 이상치 데이터 처리
copyDF = copyDF.replace('-', 0)
```

```python
# [3-4] 자료형 변환 : 48개 컬럼을 int로 형변환
d_ = dict(zip(list(copyDF.columns[2:]), ['int']*(copyDF.shape[1]-2)))
d_
```

```python
copyDF = copyDF.astype(d_)
copyDF.dtypes
```

- [4] 데이터 가공
    * 대구에서 전지역 전입 인구수 분석
    * 전출지별 - '대구광역시'
    * 전입지별 - '대구광역시', '전국' 제외

```python
# 대구광역시 데이터만 추출
mask = (copyDF['전출지별'] == '대구광역시')
cityDF = copyDF[mask]
cityDF
```

```python
# - 전출지별 컬럼 삭제
cityDF = cityDF.drop('전출지별', axis = 1)

# - 전입지별 컬럼 행 인덱스 설정
cityDF.set_index('전입지별', inplace=True)
```

```python
# - 대구광역시, 전국 행 삭제
cityDF.drop(['대구광역시', '전국'], inplace=True)
```

```python
# 가공데이터 확인
print(cityDF.dtypes, cityDF.columns, cityDF.index, sep = '\n\n')
```

```python
cityDF.index
```

```python
list_data = []
for i in cityDF.index:
    list_data.append(cityDF.loc[i].sum())

list_data
```

```python
plt.figure(figsize = (20, 10))
plt.bar(cityDF.index, list_data)
plt.title('대구 -> 타지역 인구 이동 수')
plt.xlabel('년도')
plt.ylabel('인구 이동 수')
plt.show()

for i in range(len(list_data)):
    print(f'대구광역시 -> {cityDF.index[i]} : {list_data[i]}명')
```

```python
# 광역시
metrocityDF = cityDF.loc[:'울산광역시', :]
```

```python
list_data2 = []
for i in metrocityDF.columns:
    list_data2.append(metrocityDF[i].sum())
```

```python
# 광역시 이외 도시
not_metrocityDF = cityDF.loc['세종특별자치시':, :]
```

```python
list_data3 = []
for i in not_metrocityDF.columns:
    list_data3.append(not_metrocityDF[i].sum())
```

```python
all_list1 = [metrocityDF.columns, not_metrocityDF.columns]
all_list2 = [list_data2, list_data3]
all_list3 = ['광역시', '광역시 이외 도시']
```

```python
fig, axes = plt.subplots(2, 1, figsize = (20, 10))

for i in range(2):
    axes[i].bar(all_list1[i], all_list2[i])
    axes[i].set_title(f'대구 -> {all_list3[i]} 인구 이동 수 분석')
    axes[i].set_xlabel('년도')
    axes[i].set_ylabel('인구 이동 수')
    axes[i].set_xticks(range(len(cityDF.columns))) # x축 길이만큼
    axes[i].set_xticklabels(list(cityDF.columns), rotation = 45) # x축 label에 사용할 리스트

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (20, 10))
plt.bar(cityDF.columns, cityDF.loc['서울특별시'])
plt.xticks(rotation = 45)
plt.title('대구 -> 서울 인구 이동 수')
plt.xlabel('년도')
plt.ylabel('인구 이동 수')
plt.rcParams['font.family'] = 'AppleGothic'
plt.show()
```
