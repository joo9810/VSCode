### CSV 데이터 파일 다루기
- auto_mpg 데이터 살펴보기

```python
# [1] 모듈 로딩
import pandas as pd
```

```python
# [2] 데이터 준비
file_name = '/Users/anhyojun/VSCode/K-Digital Training/Pandas/DAY04/auto_mpg.csv'
```

```python
# [3] CSV 데이터 ==> DataFrame 형태로 저장
mpgDF = pd.read_csv(file_name)
mpgDF
```

```python
print(f'데이터 형 개수 : {mpgDF.shape[0]}, 열 개수 : {mpgDF.shape[1]}')
print(f'데이터 컬럼 : \n{mpgDF.columns}')
print(f'데이터 인덱스 : \n{mpgDF.index}')
print(f'데이터 타입 : \n{mpgDF.dtypes}')
```

```python
# [5] 데이터 추출
# [5-1] 70년대 자동차 데이터만 추출
mpgDF[(mpgDF['model year'] >= 70) & (mpgDF['model year'] <= 79)]
```

```python
# [5-2] 실린더 개수가 4개 이하인 자동차 데이터만 추출
mpgDF[mpgDF['cylinders'] <= 4]
```

```python
# [5-3] 일본 차가 아닌 나머지 자동차 데이터 추출
mpgDF[~(mpgDF['origin'] == 3)]
# mpgDF[mpgDF['origin'] != 3] 이거도 됨
```

```python
# [5-4] 80년대 유럽 자동차 중에서 실린더 개수가 6개 이상인 데이터만 추출
mpgDF[(mpgDF['model year'] >= 80) & (mpgDF['model year'] <= 89) &
      (mpgDF['cylinders'] >= 6) & (mpgDF['origin'] == 2)]
```

```python
mpgDF[(mpgDF['model year'] // 10 == 8) & (mpgDF['cylinders'] >= 6) & (mpgDF['origin'] == 2)]
```

### [데이터의 정보 확인 - 통계 메서드]
- 데이터의 컬럼별 분포 및 특징 파악하기 위해서 사용
- 데이터 기본적인 분석에 활용

```python
# [1] 모듈 로딩
import pandas as pd
```

```python
# [2] 데이터 준비
file_name = '/Users/anhyojun/VSCode/K-Digital Training/Pandas/DAY04/auto_mpg.csv'
```

```python
# [3] CSV 파일 ==> DataFrame 형태로 저장
mpgDF = pd.DataFrame(pd.read_csv(file_name))
mpgDF
```

```python
# [4] 데이터 확인
# - 기본 정보 확인
mpgDF.info()
```

```python
# - 컬럼과 실제 데이터 타입 확인
mpgDF.head()
```

```python
# - 컬럼과 실제 데이터 타입 확인
print(mpgDF.head(2), mpgDF.tail(2), sep = '\n\n')
```

```python
# - 컬럼별 데이터 분포 확인
mpgDF.describe()
```

```python
# [5] 데이터 관련 통계 메서드
# - 평균
mpgDF.mean(numeric_only=True) # 수치 데이터만 계산
```

```python
# - 합계
mpgDF.sum(numeric_only=True) # 수치 데이터만 계산
```

```python
# - 중앙값
mpgDF.median(numeric_only=True)
```

```python
# - 최빈값
mpgDF.mode()
```

```python
mpgDF['weight'].value_counts() # 1985랑 2130의 빈도가 같아서 2개가 같이 나옴
```

```python
# - 고유값 : 컬럼 안에 값의 종류
mpgDF['origin'].unique(), mpgDF['origin'].value_counts()
```

```python
mpgDF['horsepower'].unique()
```

```python
# 컬럼들의 관계 정도를 수치화 나타낸것
mpgDF.corr(numeric_only=True)
```
