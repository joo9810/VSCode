## 공공데이터 과제 #1

1. 대구 기온 데이터에서 시작 연도, 마지막 연도를 입력하고 특정 월의 최고 기온 및 최저 기온의 평균값을 구하고 그래프로 표현

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
df = pd.read_csv('daegu-utf8.csv', encoding = 'utf-8-sig')
df
```

```python
def draw_low_high():
    start_year = int(input('시작 연도를 입력하세요: '))
    end_year = int(input('마지막 연도를 입력하세요: '))
    month = int(input('기온 변화를 측정할 달을 입력하세요: '))

    df['날짜'] = pd.to_datetime(df['날짜'], format = '%Y-%m-%d')

    high_temp_list = []
    low_temp_list = []

    for i in range(start_year, end_year + 1):
        high_temp_list.append(df[(df['날짜'].dt.year == i) & (df['날짜'].dt.month == month)]['최고기온'].mean())
        low_temp_list.append(df[(df['날짜'].dt.year == i) & (df['날짜'].dt.month == month)]['최저기온'].mean())

    plt.figure(figsize = (20, 5))
    plt.rcParams['font.family'] = 'AppleGothic'
    plt.plot(high_temp_list, label = '최고기온', color = 'r', marker = 's')
    plt.plot(low_temp_list, label = '최저기온', color = 'b', marker = 's')
    plt.title(f'{start_year}년부터 {end_year}까지의 {month}월의 기온 변화')
    plt.xticks(range(end_year + 1 - start_year), list(range(start_year, end_year + 1)))
    plt.legend()
    plt.show()
```

```python
draw_low_high()
```

```python
draw_low_high()
```
