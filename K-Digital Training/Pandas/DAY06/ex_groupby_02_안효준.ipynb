### [그룹화 처리]
- 주제 : 타이타닉에서 선실의 등급에 따른 생존 비율을 분석

```python
# [1] 모듈 호출
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
titanic
```

```python
titanic = titanic.drop_duplicates()
```

```python
titanic.isna().sum()
```

```python
classDF = titanic.loc[:, ['survived', 'pclass']]
classDF
```

```python
group_class = classDF.groupby('pclass')
```

```python
group_class.mean()
```
