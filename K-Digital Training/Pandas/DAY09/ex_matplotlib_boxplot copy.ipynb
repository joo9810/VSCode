### [파이 그래프]
- 컬럼(변수)의 데이터의 분포 비율을 시각화

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [2] 데이터 준비
F_DATA = r'/Users/anhyojun/VSCode/K-Digital Training/Pandas/DAY08/auto_mpg.csv'
```

```python
# [3] CSV => DataFrame으로 저장
dataDF = pd.read_csv(F_DATA)
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율
round(dataDF['origin'].value_counts() / dataDF['origin'].count(), 2)
```

```python
originSR = dataDF['origin'].value_counts()
x = list(originSR.index)
h = list(originSR.values)
x, h
```

```python
# - 자동차 생산국가에 대한 비율 시각화
freq = plt.bar(x, h, color = ['red', 'yellow', 'green'], edgecolor = ['blue'])
plt.xticks(x, ['미국', '유럽', '일본'], rotation = 45, fontsize = 15)
plt.rcParams['font.family'] = 'AppleGothic'
plt.title('자동차 생산국가에 대한 생산량')
plt.xlabel('국가')
plt.ylabel('생산량')
plt.bar_label(container = freq, labels = h, label_type = 'center')
plt.show()
```

```python
# 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels = ['미국', '유럽', '일본'], autopct = f'%.1f%%')
plt.show()
```

```python
# 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels = ['미국', '유럽', '일본'], autopct = f'%.1f%%', counterclock = False,
        startangle = 90, explode = [0.1, 0.1, 0.1])
plt.show()
```

```python
# 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels = ['미국', '유럽', '일본'], autopct = f'%.1f%%', counterclock = False,
        startangle = 90, wedgeprops = {'linewidth' : 2, 'width' : 0.7, 'edgecolor' : 'w'})
plt.show()
```
