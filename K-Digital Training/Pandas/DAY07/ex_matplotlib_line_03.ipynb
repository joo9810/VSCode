### [Matplotlib 활용 데이터 시각화]
- X축과 Y축 다양한 처리

```python
# [1] 모듈 로딩
from matplotlib import pyplot as plt
```

```python
# [2] 그래프 그리기 - 기본
y_data = [1, 3, 5, 7, 9]
x_data = ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']
plt.plot(x_data, y_data, marker = '*', color = 'red')
plt.xticks(rotation = 45)
plt.show()
```

```python
# [2] 그래프 그리기 - x축과 y축 범위 설정
y_data = [1, 3, 5, 7, 9]
x_data = ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']
plt.plot(x_data, y_data, marker = '*', color = 'red')
# plt.xlim(left = '2024-01-01', right = '2024-01-05')
plt.ylim(bottom = 0, top = 15)
plt.show()
```

```python
# [2] 그래프 그리기 - x축과 y축의 눈금(ticks) 설정
y_data = [1, 3, 5, 7, 9]
x_data = ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']
plt.plot(x_data, y_data, marker = '*', color = 'red')
# plt.xlim(left = '2024-01-01', right = '2024-01-05')
plt.ylim(bottom = 0, top = 15)
plt.xticks(rotation = 45)
plt.show()
```

```python
# [2] 그래프 그리기 - x축과 y축의 눈금(ticks) 설정
y_data = [1, 3, 5, 7, 9]
x_data = ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']
plt.figure(figsize=(7,5))       # 원하는 창의 크기
plt.plot(x_data, y_data, marker = '*', color = 'red')
# plt.xlim(left = '2024-01-01', right = '2024-01-05')
plt.ylim(bottom = 0, top = 15)
plt.xticks(rotation = 45)
plt.show()
```

```python
# [2] 그래프 그리기 - 그래프 출력 창 즉, figure를 자동 맞춤
y_data = [1, 3, 5, 7, 9]
x_data = ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']

plt.plot(x_data, y_data, marker = '*', color = 'red')
# plt.xlim(left = '2024-01-01', right = '2024-01-05')
plt.ylim(bottom = 0, top = 15)
plt.tight_layout()      # 모든 내용 출력될 수 있도록 자동 맞춤 설정
plt.show()
```

```python
# [2] 그래프 그리기 - 그래프 출력 창 즉, figure를 자동 맞춤
y_data = [1, 3, 5, 7, 9]
# x_data = ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']

plt.plot(x_data, y_data, marker = '*', color = 'red')
# plt.xlim(left = '2024-01-01', right = '2024-01-05')
plt.ylim(bottom = 0, top = 15)
plt.xticks([1, 2, 3, 4, 5])
plt.show()
```

```python
# [2] 그래프 그리기 - 그래프 출력 창 즉, figure를 자동 맞춤
y_data = [1, 3, 5, 7, 9]
# x_data = ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']

plt.plot(y_data, marker = '*', color = 'red')
# plt.xlim(left = '2024-01-01', right = '2024-01-05')
plt.ylim(bottom = 0, top = 15)
font_config = {'fontsize' : 22, 'fontweight' : 'bold'}
plt.xticks([0, 1, 2, 3, 4], ['A', 'B', 'C', 'D', 'E'], **font_config)
plt.show()
```
