```python
# 모듈 로딩
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기
df = pd.read_excel("시도별 전출입 인구수.xlsx")
df
```

```python
df.info() # 325행 50열 데이터
```

```python
# 데이터 통계량 확인
df.describe()
```

```python
# 전출지별 컬럼 결측치 처리
df[df.columns[0]] = df[df.columns[0]].ffill()
df
```

```python
# 멀티 인덱싱
column_list = list(df.columns[:2])
column_list
```

```python
df.set_index(column_list, inplace=True)
df
```

```python
# '-'를 NaN으로 변경 후 결측치 개수 확인
df.replace('-', np.NaN, inplace = True)
df.isna().sum()
```

```python
# 결측치를 모두 0으로 채우기
df.fillna(0, inplace = True)
df
```

```python
# 결측치 처리 완료
df.isna().sum()
```

## 대구 ==> 타 지역으로 이동 인구수 분석 및 시각화

```python
daegu = df.loc['대구광역시']
daegu
```

```python
# 2012~2017년 사이 대구에서 city로 인구 이동 수 만들어주는 함수
def population_move_daegu(city):
    list_data = []

    for i in daegu.columns[-6:]:
        list_data.append(daegu.loc[city, i])

    return list_data
```

```python
all_list = []

for i in list(daegu.index[1:]):
    all_list.append(population_move_daegu(i))

all_list
```

```python
# 행 개수만큼 인덱스 지정
idx = np.arange(len(daegu.columns[-6:]))

# 그래프 사이즈 지정
plt.figure(figsize = (20, 10))
for i in range(len(list(daegu.index[1:]))):
    plt.bar(idx + i*0.05, all_list[i], width = 0.05, label = list(daegu.index[1:])[i])

plt.xticks(idx, daegu.columns[-6:])
plt.rcParams['font.family'] = 'AppleGothic'
plt.legend()
plt.show()
```

```python
# 대구광역시 날리기
daegu2 = daegu.drop('대구광역시')
daegu2
```

```python
all_list2 = []

for i in list(daegu2.index[1:]):
    all_list2.append(population_move_daegu(i))

all_list2
```

```python
def display_plot(city):
    fig, axes = plt.subplots(3, 2, figsize = (20, 20))

    for i in range(3):
        for j in range(2):
            axes[i, j].bar(list(city.index[1:]),
                        list(city.loc['서울특별시':, city.columns[-6:][2*i + j]]))
            axes[i, j].set_title(f'{city.columns[-6:][2*i + j]}년') # 제목 설정
            axes[i, j].set_xlabel('도시')
            axes[i, j].set_ylabel('인구 이동 수')
            axes[i, j].set_xticks(range(len(city.index[1:])))
            axes[i, j].set_xticklabels(list(city.index[1:]), rotation=45)
```

```python
display_plot(daegu2)

plt.suptitle('2012~2017년 대구 -> 타지역 인구 이동 수', fontsize = 20)
plt.rcParams['font.family'] = 'AppleGothic'
plt.tight_layout()
plt.show()
```

## 서울 ==> 경기도 중 어느 년도가 가장 많이 갔는지 분석

```python
plt.bar(gyeonggi.columns[-6:], gyeonggi.loc['서울특별시', '2012':'2017'])
plt.xlabel('년도')
plt.ylabel('인구 이동 수')
plt.title('2012~2017년 서울 -> 경기도 인구 이동 수')
plt.show()
# 2012년에 가장 많이 이동했다.
```

## 서울 ==> 경기도 이동의 년도별 증감률 분석

```python
sg_list = list(gyeonggi.loc['서울특별시', '2012':'2017'])
sg_list
```

```python
up_down_list = []
# 서울 -> 경기 인구 이동 수 증감률
# 증감률 공식 : (최종 수치 - 최초 수치) / 최초 수치 * 100
for i in range(len(sg_list)-1):
    up_down_list.append((sg_list[i+1] - sg_list[i]) / sg_list[i] * 100)
```

```python
labels = ['2012~2013', '2013~2014', '2014~2015', '2015~2016', '2016~2017']

plt.bar(labels, up_down_list)
plt.rcParams['font.family'] = 'AppleGothic'
plt.xlabel('년도')
plt.ylabel('증감률')
plt.title('년도별 서울 -> 경기 인구 이동 수 증감률')
plt.show()

for i in range(len(labels)):
    print(f'{labels[i]} : {np.round(up_down_list[i], 2)}%')
```
